```python
import seaborn as sns
import altair as alt
```

```python
data = sns.utils.load_dataset("penguins")
```

```python
alt.Chart(data).mark_circle(size = 75, fillOpacity = 0.5).encode(
    x = alt.X('culmen_length_mm:Q', title = 'Longitud del pico (mm)'),
    y = alt.Y('body_mass_g:Q', title = 'Masa corporal (g)'),
    column = alt.Column('species:N', title = 'Especies'),
    row = alt.Row('island:N')
).properties(
    title = "Relación entre cresta del pico y masa corporal",
    height = 150,
    width = 150,
)
```

```python
barras = alt.Chart(data).mark_bar().encode(
    x = alt.X('mean(culmen_length_mm):Q'),
    y = alt.Y('island:N'),
)

etiquetas = barras.mark_text(
    align = 'left',
    dx = 3
).encode(
    text = 'island:N'
)

barras + etiquetas
```

```python
barras = alt.Chart(data).mark_bar().encode(
    x = alt.X('culmen_length_mm:Q', bin=True, title = 'Longitud del pico (mm)'),
    y = alt.Y('count()', title = 'Masa corporal (g)'),
).properties(
    title = "Relación entre cresta del pico y masa corporal"
)

linea = barras.mark_rule(color='red').encode(
    x='mean(culmen_length_mm):Q',
    size=alt.value(2)
)

text = alt.Chart(data).mark_text(
    align='left',
    baseline='middle',
    dx=7,
    dy=-135,
    size=11
).encode(
    x='mean(culmen_length_mm):Q',
    x2='mean(culmen_length_mm):T',
    text='mean(culmen_length_mm)',
    color=alt.value('#000000')
)

(barras + linea + text).properties(width=600)
```
